Se realizará un analisis EDA usando Python para examinar el nivel de obesidad de un individuo en funcion a muchos factores como por ejemplo la edad, el genero, la estatura, consumo de alcohol, etc.

La obesidad se ha convertido en una de las principales preocupaciones de salud pública a nivel mundial. Según la Organización Mundial de la Salud (OMS), en 2022, más de 1.900 millones de adultos de 18 años o más tenían sobrepeso, y de ellos, más de 650 millones eran obesos. Esta epidemia no solo afecta la calidad de vida de las personas, sino que también tiene repercusiones significativas en los sistemas de salud y la economía.

Uno de los motivos más preocupantes para estudiar la obesidad es su asociación con una serie de enfermedades crónicas. La obesidad se vincula con un aumento en la incidencia de enfermedades cardiovasculares, diabetes tipo 2, ciertos tipos de cáncer y problemas musculo-esqueléticos. De hecho, se estima que el sobrepeso y la obesidad son responsables del 44% de los casos de diabetes y del 23% de los casos de enfermedad coronaria en todo el mundo. Estos datos destacan la necesidad urgente de abordar la obesidad como un problema de salud pública.

Además, la obesidad no solo impacta la salud física, sino que también afecta la salud mental. Las personas obesas a menudo enfrentan estigmatización y discriminación, lo que puede llevar a problemas de autoestima, depresión y ansiedad. Un estudio publicado en el Journal of Health Psychology reveló que los individuos con obesidad tienen un riesgo significativamente mayor de experimentar trastornos de salud mental, lo que subraya la importancia de abordar la obesidad desde una perspectiva integral.

Otro aspecto a considerar es el costo económico asociado con la obesidad. La OMS estima que los costos directos e indirectos de la obesidad representan miles de millones de dólares cada año. Los gastos médicos relacionados con enfermedades asociadas a la obesidad aumentan la carga financiera en los sistemas de salud pública. Por ejemplo, en Estados Unidos, se calcula que la obesidad cuesta más de 147 mil millones de dólares anualmente en atención médica. Esto resalta la necesidad de invertir en la investigación y prevención de la obesidad para reducir estos costos.

El estudio de la obesidad también es crucial para entender su complejidad. La obesidad es el resultado de una interacción de factores genéticos, ambientales, sociales y comportamentales. Investigaciones recientes han mostrado que factores como el acceso a alimentos saludables, la actividad física, y el entorno social y económico juegan un papel fundamental en el desarrollo de la obesidad. Por lo tanto, una comprensión más profunda de estos factores puede guiar el desarrollo de políticas efectivas y programas de intervención.

A nivel global, la obesidad no es un problema uniforme; varía significativamente entre diferentes regiones y grupos demográficos. Por ejemplo, la prevalencia de la obesidad ha ido en aumento en países de ingresos medios y bajos, lo que plantea nuevos desafíos. Según un informe de la OMS, en 2022, el 39% de los adultos en América Latina y el Caribe tenían sobrepeso u obesidad. Esto subraya la necesidad de un enfoque específico y culturalmente relevante para abordar el problema.

Por último, el estudio de la obesidad es vital para promover un estilo de vida saludable en la población. La educación sobre nutrición y la promoción de la actividad física son componentes esenciales para prevenir la obesidad. Los programas de intervención que han mostrado éxito combinan estrategias educativas, políticas públicas y cambios en el entorno para facilitar estilos de vida más saludables.

En conclusión, el estudio de la obesidad es fundamental no solo por sus implicaciones para la salud física y mental, sino también por su impacto económico y social. Comprender los diversos factores que contribuyen a la obesidad puede guiar la formulación de políticas efectivas y la creación de programas de intervención. La prevención y el tratamiento de la obesidad son esenciales para mejorar la salud pública y reducir la carga económica asociada a esta condición.

##3.Inspeccion de datos y limpieza

El archivo contiene 17 variables y 2111 observaciones. Las variables son las siguientes:

Gender: Género de la persona (object).

Age: Edad de la persona (float64).

Height: Altura en metros (float64).

Weight: Peso en kilogramos (float64).

family_history_with_overweight: Indica si hay historia familiar de sobrepeso (object: "yes" o "no").

FAVC: Indica si la persona consume alimentos con alto contenido calórico con frecuencia (object: "yes" o "no").

FCVC: Frecuencia de consumo de vegetales (float64).

NCP: Número de comidas principales por día (float64).

CAEC: Frecuencia de consumo de alimentos entre comidas (object: valores como "Sometimes").

SMOKE: Indica si la persona fuma (object: "yes" o "no").

CH2O: Consumo de agua diario en litros (float64).

SCC: Indica si la persona monitorea su ingesta de calorías (object: "yes" o "no").

FAF: Frecuencia de actividad física en horas por semana (float64).

TUE: Tiempo usando dispositivos electrónicos en horas por día (float64).

CALC: Frecuencia de consumo de alcohol (object: valores como "no" o "Sometimes").

MTRANS: Modo de transporte principal (object: valores como "Public_Transportation", "Walking").

NObeyesdad: Nivel de obesidad según una clasificación (object: valores como "Normal_Weight", "Overweight_Level_I").
